```python
#Simple Linear Regression having line equation in the form of y=mx+c
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Step 2: Generate some sample data
# X: independent variable (feature), y: dependent variable (target)
np.random.seed(0)  # for reproducibility
X = 2 * np.random.rand(500, 1)   #  random points for X
y = 4 * X + 5 + np.random.randn(500,1) #equation = y=mx+c + noise

# Step 3: Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)

# Step 4: Get model parameters
intercept = model.intercept_[0]
slope = model.coef_[0][0]

# Step 5: Make predictions
y_pred = model.predict(X)
# Step 6: Evaluate the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

# Step 7: Print results
print("Linear Regression Equation: y = {:.3f} + {:.7f}*X".format(intercept, slope))
print("Mean Squared Error:", round(mse, 2))
print("R-squared (RÂ²):", round(r2, 2))

# Step 8: Plot the data and regression line
plt.figure(figsize=(15, 10))
plt.scatter(X, y, color='black', label='Actual Data')
plt.plot(X, y_pred, color='red', linewidth=4, label='Regression line')
plt.xlabel("X - Independent Variable")
plt.ylabel("y - Dependent Variable")
plt.title("Linear Regression Example")
plt.legend()
plt.show()
```

```python
# Simple Linear Regression were we split data in train test and Cross-Validation
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error, r2_score

#Generate Random Data for train test and validate our model
np.random.seed(0)  # For reproducibility
X = np.random.rand(500, 1) * 5   # Feature values 0-5
y = 4 * X + 5 + np.random.randn(500,1)  # y = 4X + 5 + noise

# Step 3: Split Dataset into Train and Test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=50)

# Step 4: Initialize Linear Regression Model
model = LinearRegression()

# Step 5: Cross-Validation on Training Data
cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='r2')
print("Cross-Validation RÂ² Scores:", cv_scores)
print("Average CV RÂ² Score:", cv_scores.mean())

# Step 6: Train the Model on Training Data
model.fit(X_train, y_train)

# Step 7: Get Model Parameters
intercept = model.intercept_[0]
slope = model.coef_[0][0]
print(f"\nLinear Regression Equation: y = {intercept:.3f} + {slope:.3f}*X")

# Step 8: Predict on Test Data
y_pred = model.predict(X_test)

# Step 9: Evaluate Model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print("Mean Squared Error on Test Data:", round(mse, 2))
print("RÂ² Score on Test Data:", round(r2, 2))

# Step 10: Visualize Training Data, True Line, and Regression Line
plt.figure(figsize=(15, 10))
plt.scatter(X, y, color='black', label='Actual Data')
plt.scatter(X_test, y_pred, color='red', label='Predicted Points (Test Data)')
plt.plot(X, model.predict(X), color='orange', linewidth=2, label='Regression Line')
plt.xlabel("X - Independent Variable")
plt.ylabel("y - Dependent Variable")
plt.title("Linear Regression were we split data in train test and Cross-Validation")
plt.legend()
plt.show()
```
